```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# 웹 드라이버 초기화
driver = webdriver.Chrome(executable_path="chromedriver")
driver.get("https://www.youtube.com")
time.sleep(1)

searchValues = []
print("검색어를 입력해주세요")
print("마지막은 엔터만 입력해주세요")
while True:
    searchValue = input("검색어 입력 : ")
    if searchValue == "":
        break
    searchValues.append(searchValue)

youtube_list = []

for searchValue in searchValues:
    searchInput = driver.find_element(
        by="css selector", value="input[name='search_query']")
    searchInput.clear()
    searchInput.send_keys(searchValue)
    searchInput.send_keys(Keys.RETURN)
    time.sleep(3)  # 검색 결과 로딩을 기다리기 위한 3초 대기

    youtube_items = driver.find_elements(
        by="css selector", value="div#contents ytd-video-renderer")
    youtube_items_list = []

    for item in youtube_items[:10]:
        title = item.find_element(
            by="css selector", value="a#video-title").text
        hits = item.find_element(
            by="css selector", value="span.style-scope.ytd-video-meta-block").text
        item_dict = {
            "제목(title)": title,
            "조회수(hits)": hits,
        }
        youtube_items_list.append(item_dict)

    df = pd.DataFrame(youtube_items_list)
    youtube_list.append(df)

# 모든 검색 결과를 하나의 DataFrame으로 병합
result_df = pd.concat(youtube_list, ignore_index=True)
print(result_df)
```
